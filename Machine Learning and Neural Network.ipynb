### MACHINE LEARNING AND NEURAL NETWORK

### Simple Linear Regression

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

#### DATA: Generating the data via random input to train on.

```python
observations =1000 # this is arbitrary number. You can try with different number of observations.

xs=np.random.uniform(low=-10, high=10, size =(observations,1))
zs=np.random.uniform(-10,10,(observations,1))

inputs =np.column_stack((xs,zs))

print(inputs.shape)
```

#### CREATE THE TARGETS WE WILL AIM AT

```python
noise= np.random.uniform(-1,1,(observations,1))

targets= 2*xs - 3*zs + 5 + noise

print(targets.shape)
```

#### DATA PLOT

The goal here is to see that there is a strong trend that the model should learn to reproduce.

```python
targets = targets.reshape(observations,)
fig =plt.figure()
ax=fig.add_subplot(111, projection ='3d')
ax.plot(xs, zs, targets)
ax.set_xlabel('xs')
ax.set_ylabel('zs')
ax.set_zlabel('Targets')
ax.view_init(azim=100)
plt.show()
targets=targets.reshape(observations,1)
```

For the linear regression Y=xw + b, our output will try to find the values of w and b so the output Y are closest to the targets. 

#### DECLARE VARIABLES

```python
init_range = 0.1

weights =np.random.uniform(-init_range, init_range, size= (2,1))

biases =np.random.uniform(-init_range, init_range, size =1)

print(weights)
print(biases)
```

The top two values are the weights while the bottom value is the biases.

#### SET A LEARNING RATE

```python
learning_rate =0.02
```

##### TRAINING THE MODEL USING GRADIANT DESCENT METHODOLOGY

```python
for i in range(100):
    outputs =np.dot(inputs, weights) + biases
    deltas = outputs - targets
    
    loss =np.sum(deltas ** 2) /2 /observations
    
    print(loss)
    
    deltas_scaled =deltas /observations
    
    weights= weights - learning_rate * np.dot(inputs.T,deltas_scaled)
    biases =biases -learning_rate * np.sum(deltas_scaled)
```

#### PRINT WEIGHTS AND BIASES AND SEE IF WE HAVE WORKED CORRECTLY

```python
print(weights, biases)
```

##### PLOT LAST OUTPUTS vs TARGETS

Since they are the last ones at the end of the traing, they represent the final model accuracy. The closer this plot is to a 45 degree line, the closer taget and output values are.

```python
plt.plot(outputs, targets)
plt.xlabel('Outputs')
plt.ylabel('Targets')
plt.show()
```
