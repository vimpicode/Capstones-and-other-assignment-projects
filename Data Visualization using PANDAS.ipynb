### DATA VISUALIZATION USING PANDAS

In this small exercise we use some of the commonly used functions of Data Visualization in Pandas to display different plots using mainly Pandas but also incorporating functions from other libraries such as Matplotlib and Seaborn

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
df3 = pd.read_csv('df3')
%matplotlib inline
```

```python
df3.info()
```

```python
df3.head()
```

### Create a scatter plot of b vs a.

```python
df3.plot.scatter(x='a', y='b', color= 'red', s=50, figsize=(12,3))
```

#### Plot a histogram of the 'a' column.

```python
df3['a'].plot.hist(color='blue')
```

###### Change the above histogram with 25 bins, and red color.

```python
df3['a'].plot.hist(bins=25, color='red', alpha=.5)
```

#### Create a boxplot to compare columns 'a' vs. 'b'.

```python
df3[['a','b']].plot.box()
```

#### Use kde to plot a graph for column 'd'.

```python
df3['d'].plot.kde()
```

##### In the above kde plot, change the linewidht, line style and color 

```python
df3['d'].plot.kde(lw=3, ls='--', color='red')
```

##### Create an area plot that displays all columns

```python

df3[0:30].plot.area(stacked=True, alpha=.9)
```

###### Let's display the legend outside of the diagram instead of making it overlap as it is displayed above

```python
f = plt.figure()
df3[0:30].plot.area(alpha=0.4,ax=f.gca())
plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))
plt.show()
```
